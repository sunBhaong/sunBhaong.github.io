1.时间序列生成

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

(1)	生成今天的时间戳

```python
t1 = pd.datetime.now() 
pd.to_datetime(t1)
```

(2)	生成2016年7月1日开始，长度为10的时间索引序列，时间间隔为3天。

```python
pd.date_range(start='2016/7/1',freq='3D',periods=10)
```

(3)	生成从2021年5月25日开始，间隔为12小时，长度为10的时间索引序列。

```python
pd.date_range(start='2021/5/25',periods=10,freq='12H')
```

(4)	生成2021年6月的时期对象

```python
pd.Period('2021/6')
```

(5)	生成2021年1月到6月的时期索引序列

```python
pd.date_range(start='2021/1/1',end='2021/6/30',freq='1D')
```
